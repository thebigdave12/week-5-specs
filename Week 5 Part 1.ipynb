### 2. In your own words, summarize the following concepts as they relate to visual perception:

    Order - Because there is no agreed standard for order in data visualization, one needs to create an efficient way to interpret graphs (everyone will read differently, sporadic). Thus, one should lead and guide the viewer in the way that you desire through specific design choices. 

    Hierarchy - Draw the viewer's focus to the most meaningful aspect of the data visualization. As data analysts we are trying to guide their understanding not distract the viewers. We can do this through color, size, contrast alignment, repetition etc... Design choices will allow us to tell a story that leads to action. 

    Relationships - Order in which the data is shown is crucial because the audience will begin to develop a story in their head. We have to remember that correlation is not causation so must be very decisive with what is placed into a data visualization. If we present something there must be an actual relationship. 

    Convention - Although there isn't a specific order, there are expected conventions when building a visualization. For example, the title, data point, axis, and legend all have a specific place they live and should be formatted in a data visualization.
